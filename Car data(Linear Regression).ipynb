### Building ML model for prediction

```python
# to split the data into trainning and testing data sets
from sklearn.model_selection import train_test_split
```

```python
dfcd1.columns
```

```python
dfcd1.head()
```

```python
# Make the column with non integer value to integer value so it can be used in a ML model
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
dfcd1['fuel-c']=lb.fit_transform(dfcd1['fuel'])
```

```python
lb=LabelEncoder()
dfcd1['transmission']=lb.fit_transform(dfcd1['transmission'])
```

```python
lb=LabelEncoder()
dfcd1['Company']=lb.fit_transform(dfcd1['Company'])
```

```python
dfcd1['seller_type'].unique()
```

```python
lb=LabelEncoder()
dfcd1['seller_type']=lb.fit_transform(dfcd1['seller_type'])
```

```python
dfcd1.head()
```

```python
X=dfcd1[['year', 'km_driven', 'fuel-c', 'seller_type',
       'transmission', 'Company']]
y=dfcd1['selling_price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=0)
```

```python
X.head()
```

```python
# metrics for error/precision calculation of the model
# Linear regression model for predicting 
from sklearn.metrics import mean_absolute_error
from sklearn.linear_model import LinearRegression
```

```python
li=LinearRegression()
```

```python
# fit the model with training and testing data
li.fit(X_train,y_train)
```

```python
# predict the test data and save it in a variable name 'pred' which will go in a metrics to compare with the actual value
perd=li.predict(X_test)
```

```python
mean_absolute_error(y_test,perd)
```
